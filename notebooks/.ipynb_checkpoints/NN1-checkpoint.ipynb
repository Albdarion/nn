```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def linear(x):
    return x
```

```python
class Layer:
    def __init__(self, p_num_nodes_in, p_num_nodes_out, p_activation):
        self.num_nodes_in = p_num_nodes_in
        self.num_nodes_out = p_num_nodes_out
        self.activation = p_activation
        
        return
    
    def set_weights(self, p_weights):
        self.weights = p_weights
        
        return
    
    def set_biases(self, p_biases):
        self.biases = p_biases
        
        return
    
    def forward(self, p_input):
        return self.activation(np.dot(p_input, self.weights) + np.dot(np.ones((p_input.shape[0], 1)), self.biases))
```

```python
class NeuralNetwork:    
    def __init__(self, p_layers, p_weights, p_biases, p_activation, p_output_activation):        
        self.layers = []
        
        for i in range(len(p_layers) - 1):
            func = p_activation
            if i == len(p_layers) - 2:
                func = p_output_activation
                
            layer = Layer(p_layers[i], p_layers[i + 1], func)
            layer.set_weights(p_weights[i])
            layer.set_biases(p_biases[i])
            
            self.layers.append(layer)
            
    
    def predict(self, p_input):        
        for layer in self.layers:
            p_input = layer.forward(p_input)
            
        return p_input
```

# 1. square-simple dataset

```python
square_simple_train = pd.read_csv('../data/regression/square-simple-training.csv', index_col=0)
square_simple_test = pd.read_csv('../data/regression/square-simple-test.csv', index_col=0)
```

```python
square_simple_X_train = square_simple_train.drop(columns='y')
square_simple_y_train = square_simple_train['y']

square_simple_X_test = square_simple_test.drop(columns='y')
square_simple_y_test = square_simple_test['y']

sq_sim_X_train = square_simple_X_train.to_numpy().reshape((square_simple_X_train.shape[0], 1))
sq_sim_y_train = square_simple_y_train.to_numpy().reshape((square_simple_y_train.shape[0], 1))

sq_sim_X_test = square_simple_X_test.to_numpy().reshape((square_simple_X_test.shape[0], 1))
sq_sim_y_test = square_simple_y_test.to_numpy().reshape((square_simple_y_test.shape[0], 1))
```
