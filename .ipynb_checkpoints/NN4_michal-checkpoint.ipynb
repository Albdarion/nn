```python
def F_score(Y,y_pred):
    classes = np.argmax(y_pred, axis = 1)
    y_pred = np.zeros(y.shape)
    y_pred[np.arange(len(y_pred)),classes] = 1
    y_pred = y_pred.astype(int).transpose()
    Y = Y.astype(int).transpose()
    result = np.zeros(len(Y))
    for i in range(len(Y)):
        pred_positives = np.count_nonzero(y_pred[i])
        positives = np.count_nonzero(Y[i])
        tp = np.count_nonzero(y_pred[i] & Y[i])
        if tp == 0:
            result[i]= 0
        else:
            result[i]= 2/(pred_positives/tp + positives/tp)
    return result
```

```python
def Softmax(x):
    s = np.exp(x)
    p = np.sum(s,axis = 1, keepdims=True)
    return s/p
```

```python
import warnings
def sigmo(x):
    warnings.filterwarnings('ignore')
    return 1/(1+np.exp(-x))
```

```python
def sigmo_der(x):
    s = sigmo(x)
    return s*(1-s)
```

```python
def one_hot(categories):
    unique_categories = np.unique(categories)
    one_hot_encoding = np.zeros((len(categories), len(unique_categories)))

    for i, category in enumerate(categories):
        index = np.where(unique_categories == category)[0][0]
        one_hot_encoding[i, index] = 1

    return(one_hot_encoding)
```

```python
def read(path):
    D = pd.read_csv(path)
    x = np.array(D.x)
    y = np.array(D.y)
    c = np.array(D.c)

    return np.array((x,y)).transpose(),c.astype(int)
```

```python
class NeuralNetwork(object):
    def __init__(self,layers,fun=sigmo,fun_der=sigmo_der, cost_fun = F_score, end_fun = Softmax):
        outs = layers[1:]
        ins = layers[:-1]
        self.end_fun =end_fun
        self.w = [np.random.normal(0,np.sqrt(2/(insi+outi)),size = (insi,outi)) for (insi,outi) in zip(ins,outs)]
        self.b = [np.zeros(outi) for outi in outs]
        self.fun = fun
        self.fun_der = fun_der
        self.cost_fun = cost_fun
        #wagi wchodzące w ity neuron
        self.a = [None]*len(self.w)
        #kombinacje liniowe w itym neuronie 
        self.z = [None]*len(self.w)
        self.meanx = 0
        self.meany = 0
        self.stdx = 1
        self.stdy = 1
        self.Vb = [np.zeros(bi.shape) for bi in self.b]
        self.Mb = [np.zeros(bi.shape) for bi in self.b]
        self.Vw = [np.zeros(wi.shape) for wi in self.w]
        self.Mw = [np.zeros(wi.shape) for wi in self.w]

    def set_normals(self,x,y):
        self.meanx = np.mean(x)
        self.stdx = np.std(x)
        self.meany = np.mean(y)
        self.stdy = np.std(y)
        
    def normalize(self,x):
        return (x-self.meanx)/self.stdx

    
    def fit(self,x):
        #x jako pierwsza warstwa, wiersz to jedna obserwacja 
        n = len(self.w)-1
        l = x
        self.z[0] = l
        x = self.normalize(x)
        #rekurencyjna pętla przechodząca po warstwach
        for i in range(n):
            self.a[i]=l
            l = l@self.w[i]+self.b[i]
            self.z[i+1]= l
            l = self.fun(l)
        
        self.a[-1]=l
        #doprowadzenie do prostej listy
        ans = self.end_fun(l@self.w[-1]+self.b[-1])
        return ans
    
    def gradient(self,x,y):
        n = len(self.w)
        e = [np.zeros(b.shape) for b in self.b]
        g = [np.zeros(w.shape) for w in self.w]
        #fit zapamietuje wartosci wchodzace i te po kombinacjach liniowych
        y_pred = self.fit(x)

        #delta = np.row_stack(-y*np.log(y_pred+0.000001)- (1-y)*np.log(1-y_pred+0.000001))
        delta = (y_pred-y)/len(x)
        for i in range(1,n+1):
                e[-i] = np.sum(delta,axis = 0)
                g[-i] = (self.a[-i].transpose())@delta
                delta = (delta@(self.w[-i].transpose()))*(self.fun_der(self.z[-i]))
        return e, g
    
    def train(self,x,y,batchsize, learn_ratio, epochs, random = False, method = "SGD", beta = 0.9):
        n = len(x)
        self.set_normals(x,y)
        cost= [None]*epochs
        for i in range(epochs):
            indexes = list(range(n))
            if random:
                np.random.shuffle(indexes)
            for k in range(0,n,batchsize):
                #bierzemy batchsize wybranych indeksow
                use = indexes[k:k+batchsize]
                #liczymy gradienty dla tych obserwacji i aktualnych parametrów
                Gb, Gw = self.gradient(x[use],y[use])
                if method == "RMSProp":
                    Gb, Gw = self.RMSProp(Gb,Gw,beta)
                elif method == "Momentum":
                    Gb, Gw = self.Momentum(Gb,Gw,beta)
                #aktualizacja parametrów
                self.w = [wi- learn_ratio*gi for (wi,gi) in zip(self.w,Gw)]
                self.b = [bi- learn_ratio*gi for (bi,gi) in zip(self.b,Gb)]
            #zapis mse po użyciu pełnego epoch danych
            cost[i] = self.cost_fun(y,self.fit(x))
        return np.array(cost).transpose()
    
    def Momentum(self,Gb,Gw,beta):
        self.Vb = [beta*vbi + gbi for (vbi,gbi) in zip(self.Vb,Gb)]
        self.Vw = [beta*vwi + gwi for (vwi,gwi) in zip(self.Vw,Gw)]
        return self.Vb, self.Vw

    def RMSProp(self,Gb,Gw,beta):
        self.Mb = [beta*mbi + (1-beta)*np.square(gbi) for (mbi,gbi) in zip(self.Mb,Gb)]
        self.Mw = [beta*mwi + (1-beta)*np.square(gwi) for (mwi,gwi) in zip(self.Mw,Gw)]
        #aktualizacja parametrów
        Gw = [np.divide(gwi,np.sqrt(mwi)+0.00001) for (mwi,gwi) in zip(self.Mw,Gw)]
        Gb = [np.divide(gbi,np.sqrt(mbi)+0.00001) for (mbi,gbi) in zip(self.Mb,Gb)]
        return Gb,Gw
```

```python
def compare(f_scoreSoftmax, f_scoreSigmoid):
    plt.scatter(list(range(len(f_scoreSigmoid[0]))),np.mean(f_scoreSoftmax,axis=0), label ="Średnie F-score dla Softmax")
    plt.scatter(list(range(len(f_scoreSigmoid[0]))),np.mean(f_scoreSigmoid,axis=0), label ="Średnie F-score dla Sigmoid")
    plt.ylabel("F-score")
    plt.xlabel("epochs")
    plt.legend()
    plt.show()
```

```python
x,y = read("classification/easy-training.csv")
y = one_hot(y)
model1Softmax = NeuralNetwork([2,5,2])
model1Sigmoid = NeuralNetwork([2,5,2], end_fun=sigmo)
f_scoreSoft =model1Softmax.train(x,y,5,0.001,100, method="RMSProp")
f_scoreSigmo =model1Sigmoid.train(x,y,5,0.001,100, method="RMSProp")
compare(f_scoreSoft,f_scoreSigmo)
```

```python
x,y = read("classification/easy-training.csv")
y = one_hot(y)
epochs = 100
model1 = NeuralNetwork([2,2])
f_score = model1.train(x,y, 5, 0.001,epochs, method ="RMSProp")
print("F_score dla trainu:" ,model1.cost_fun(y,model1.fit(x)))
x,y =read("classification/easy-test.csv")
y = one_hot(y)
print("F_score dla testu:" ,model1.cost_fun(y,model1.fit(x)))
plot(f_score)
```
